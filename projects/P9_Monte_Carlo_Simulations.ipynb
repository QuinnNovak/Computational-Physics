# Monte Carlo Estimation of π — Narrative and Analysis

---

## Working Code and Computational Approach

In this project, I implemented two Monte Carlo simulations to estimate the value of π:

1. **Dartboard (area–ratio) method**
2. **Buffon's Needle method**

### Dartboard Method
The dartboard simulation:

- Generates \(N\) random points \((x,y)\) uniformly inside $$[-1,1]\times[-1,1]$$.
- Classifies points as inside the unit circle if  
  $$ x^2 + y^2 \le 1 $$
- Estimates π using  
  $$ \hat{\pi} = 4 \cdot \frac{N_{\text{in}}}{N} $$
- Produces:
  - A scatter plot showing darts inside/outside the circle  
  - A convergence plot $$ \hat{\pi}(N)$$

### Buffon's Needle Method
The needle simulation:

- Draws random angle $$ \theta \sim U(0,\pi) $$
- Draws random center offset $$ x\sim U(0,d/2) $$
- Applies the crossing condition  
  $$ \frac{L}{2}\sin\theta \ge x $$
- Uses Buffon’s estimator  
  $$ \hat{\pi} = \frac{2N}{H} $$
  where \(H\) is the number of crossing events.

---

## Mathematical Principles and Underlying Physics

### 1. Dartboard Method: Geometric Probability

Square of side length 2:
$$ A_{\text{square}} = 4 $$

Circle of radius 1:
$$ A_{\text{circle}} = \pi $$

Probability of landing inside:
$$ P(\text{inside}) = \frac{\pi}{4} $$

Empirical fraction:
$$ \frac{N_{\text{in}}}{N} $$

Law of Large Numbers:
$$ \frac{N_{\text{in}}}{N} \to \frac{\pi}{4} $$

Estimator:
$$ \hat{\pi} = 4 \cdot \frac{N_{\text{in}}}{N} $$

Monte Carlo uncertainty:
$$ \text{error} \sim \frac{1}{\sqrt{N}} $$

---

### 2. Buffon’s Needle: Angular Probability

Crossing condition:
$$ \frac{L}{2}\sin\theta \ge x $$

For \(L=d=1\):
$$ P(\text{cross}\mid\theta) = \sin\theta $$

Average over all angles:
$$ P(\text{cross}) = \frac{1}{\pi} \int_0^\pi \sin\theta \, d\theta = \frac{2}{\pi} $$

Thus:
$$ \hat{\pi} = \frac{2N}{H} $$

```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi_darts(N, seed=None, make_scatter=False):
    rng = np.random.default_rng(seed)
    x = rng.uniform(-1.0, 1.0, N)
    y = rng.uniform(-1.0, 1.0, N)
    r2 = x**2 + y**2
    inside = r2 <= 1.0
    pi_hat = 4.0 * inside.mean()

    if make_scatter:
        theta = np.linspace(0, 2*np.pi, 400)
        plt.plot(np.cos(theta), np.sin(theta))               # circle
        plt.scatter(x[inside], y[inside], s=5)
        plt.scatter(x[~inside], y[~inside], s=5, marker='x')
        plt.gca().set_aspect('equal', 'box')
        plt.title(f"Dartboard: N={N}, pi≈{pi_hat:.5f}")
        plt.grid(True)
        plt.show()

    return pi_hat

def convergence_plot_darts(max_N=50000, step=500, seed=0):
    rng = np.random.default_rng(seed)
    x = rng.uniform(-1.0, 1.0, max_N)
    y = rng.uniform(-1.0, 1.0, max_N)
    r2 = x**2 + y**2
    inside = r2 <= 1.0

    Ns = np.arange(step, max_N+1, step)
    pi_vals = []

    for N in Ns:
        pi_vals.append(4 * inside[:N].mean())

    plt.plot(Ns, pi_vals)
    plt.axhline(np.pi)
    plt.xlabel("Number of darts")
    plt.ylabel("Pi estimate")
    plt.title("Convergence of Pi (Dartboard)")
    plt.grid(True)
    plt.show()

    return Ns, np.array(pi_vals)


# ---- RUN THE FUNCTIONS ----
estimate_pi_darts(2000, seed=1, make_scatter=True)
convergence_plot_darts()

def buffon_pi(N, L=1.0, d=1.0, seed=None):
    rng = np.random.default_rng(seed)
    theta = rng.uniform(0, np.pi, N)
    x_center = rng.uniform(0, d, N)

    # Distance from center to crossing
    y = (L/2) * np.sin(theta)

    # Crosses a line if center is within y of the boundary
    crosses = (x_center < y) | (x_center > d - y)
    H = crosses.sum()

    if H == 0:
        return np.nan, crosses

    pi_hat = (2 * L * N) / (H * d)
    return pi_hat, crosses

def example_plot_buffon(N=200, L=1.0, d=1.0, seed=0):
    rng = np.random.default_rng(seed)
    theta = rng.uniform(0, np.pi, N)
    x0 = rng.uniform(0, 5, N)
    y0 = rng.uniform(0, 5, N)

    dx = 0.5 * L * np.cos(theta)
    dy = 0.5 * L * np.sin(theta)

    x1 = x0 - dx; x2 = x0 + dx
    y1 = y0 - dy; y2 = y0 + dy

    # Lines
    for xl in np.arange(0, 6, 1):
        plt.axvline(xl, color='gray')

    # Needles
    for i in range(N):
        plt.plot([x1[i], x2[i]], [y1[i], y2[i]])

    pi_hat, _ = buffon_pi(N, L, d, seed)
    plt.title(f"Buffon's Needle (N={N}), pi≈{pi_hat:.5f}")
    plt.gca().set_aspect('equal', 'box')
    plt.grid(True)
    plt.show()


def convergence_plot_buffon(max_N=50000, step=500, L=1.0, d=1.0, seed=1):
    Ns = []
    pi_vals = []

    for N in range(step, max_N+1, step):
        pi_hat, _ = buffon_pi(N, L, d, seed)
        Ns.append(N)
        pi_vals.append(pi_hat)

    Ns = np.array(Ns)
    pi_vals = np.array(pi_vals)

    plt.plot(Ns, pi_vals)
    plt.axhline(np.pi)
    plt.xlabel("Number of needles")
    plt.ylabel("Pi estimate")
    plt.title("Convergence of Pi (Buffon's Needle)")
    plt.grid(True)
    plt.show()

    return Ns, pi_vals


# ---- RUN THE FUNCTIONS ----
example_plot_buffon()
convergence_plot_buffon()
```
